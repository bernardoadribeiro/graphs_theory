## Representação para verificar a planaridade do grafo G




### Referências: 
- Biblioteca Python: [NetworkX](https://networkx.org)
- 



```python
import networkx as nx
```

```python
# Grapfo K_5 -> Não planar
K_5 = nx.complete_graph(5)
print("É planar >>> ", nx.is_planar(K_5))
nx.draw(K_5, with_labels=True, font_weight='bold')
```

```python
# Grafo K_3,3 -> Não planar
K_3_3 = nx.complete_bipartite_graph(3, 3)
print("É planar >>> ", nx.is_planar(K_3_3))
nx.draw(K_3_3, with_labels=True, font_weight='bold')
```

```python
# Grafo K_4 -> Planar
K_4 = nx.complete_graph(4)
print("É planar >>> ", nx.is_planar(K_4))
nx.draw_planar(K_4, with_labels=True, font_weight='bold')
```
