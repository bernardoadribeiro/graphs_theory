## Representação para verificar a planaridade do grafo G
Um grafo é considerado planar se ele puder ser desenhado sem que qualquer uma das arestas (edges) se cruzem.

De forma bem resumida, um grafo é considerado plano/planar, quando representado graficamente (Desenho) as suas arestas não se cruzam.
- Caso um Grafo G:
   - Consiga ser desenhado sem o cruzamento de arestas: Ele É Planar (Grafo Planar)
   - Não Consiga ser desenhado sem o cruzamento de arestas: Ele NÃO É Planar (Grafo Não-Planar)

Um grafo planar divide o plano em **Regiões (ou Faces)**

### Definições
**Grafos Completos e Bi-Partidos**
- **Completos**: É um grafo em que **todos** os vértices estão conectados à **todos** os outros vértices.
- **Bi-Partidos**: Um grafo é dito ser bipartido quando seu conjunto de vértices V puder ser particionado em dois subconjuntos V1 e V2, tais que toda aresta de G une um vértice de V1 a outro de V2.
- **Bi-Partidos completo**: Ele é completo quando **todos** os vértices de uma partição estão ligados a **todos** os vértices da outra partição.


### Formas de verificar a planaridade de um grafo:
- **Desenho gráfico**: É possível desenhar um grafo e verificar se as arestas se sobrepõem;
- **Teorema de Kuratowski**: é planar se e somente se ele não contém um subgrafo homeomórfico a K_{5}​ (grafo completo com cinco vértices) ou a K_{3,3}​ (grafo bipartido completo com seis vértices). Portanto, você pode verificar se um grafo contém um desses subgrafos para determinar se ele é planar ou não.
- **Fórmula de Euler**: Determina a quantidade de faces (f) de um grafo baseado no número de vértices (n) e no número de arestas (m)
                                                      `f = m - n + 2`
   
   Se G é um grafo planar e conexo, com m > 1, então a fórmula abaixo se aplica:
                                                        ` m ≤ 3n - 6`

   Todo gráfico planar atende à essa condição, **mas há grafos não-planares que também atendem**.

### Grafos Não Planares

```python
# Grafo K_5 -> Não planar
# O grafo K_{5} é o grafo não-planar com o menor número de vértices

K_5 = nx.complete_graph(5)
print(K_5)
print("É planar >>> ", nx.is_planar(K_5))

nx.draw(K_5, with_labels=True, font_weight='bold')
```

```python
# Grafo K_{3,3} -> Não planar
# O Grafo K_{3,3} é o grafo não planar com o menor número de arestas

K_3_3 = nx.complete_bipartite_graph(3, 3)
print("É planar >>> ", nx.is_planar(K_3_3))

nx.draw(K_3_3, with_labels=True, font_weight='bold')
```

### Grafos Planares

```python
# Grafo K_4 -> Planar

K_4 = nx.complete_graph(4)
print("É planar >>> ", nx.is_planar(K_4))

nx.draw_planar(K_4, with_labels=True, font_weight='bold')
```

```python
# Grafo K_{3,2} -> Planar

K_3_2 = nx.complete_bipartite_graph(3, 2)
print("É planar >>> ", nx.is_planar(K_3_2))

nx.draw_planar(K_3_2, with_labels=True, font_weight='bold')
```
